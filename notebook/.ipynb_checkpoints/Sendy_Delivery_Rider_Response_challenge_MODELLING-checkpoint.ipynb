## PART D: MODEL BUILDING AND EVALUATION

It's time to build the final model to predict whether a rider will **accept**, **decline** or **ignore** an order sent to them.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
# %Loading clean dataset
df = pd.read_csv("../data/Final_data_0.csv")
df.head()
```

#### FEATURE SELECTION

```python
# %spliting dependent and independent features
features_cols = df.drop(["target","vendor_type"],1)
cols = features_cols.columns
target=df["target"]
```

```python
df[cols].shape , target.shape
```

#### MODELING

```python
# %Let's set seed for reproducibility of our final model

SEED = 360 #you can set any number
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df[cols], target, test_size = 0.2, random_state = SEED)
```

##### CATBOOST

```python
from catboost import Pool, CatBoostClassifier
from sklearn.metrics import accuracy_score, confusion_matrix , plot_confusion_matrix
```

```python
# %selecting categorical features
# cat_features = [0]
X = df[cols]
categorical_features_indices = np.where(X.dtypes != np.float)[0]

# %training set
train_dataset = Pool(data=X_train,
                     label=y_train,
                     cat_features=categorical_features_indices)

# %evaluation set
eval_dataset = Pool(data=X_test,
                    label=y_test,
                    cat_features=categorical_features_indices)

# Initialize CatBoostClassifier
cat_model = CatBoostClassifier(iterations=1000, eval_metric= "Accuracy", early_stopping_rounds = 100,loss_function='MultiClass')

# Fit model
cat_model.fit(train_dataset)

# Get predicted classes
preds_class = cat_model.predict(eval_dataset)

# Check score
accuracy_score(y_test, preds_class)
```

#### VISUALIZE & INTERPRETATION OF THE PREDICTIONS

This will help to understand how catboost model making predictions after training.Through interpretations will helps to figure out when the model will work well and when it wonâ€™t work well. 

**Recall:** our target is to build something which will help to improve customer service on sendy platform

```python
# %Let's use confunsion matrix plot to Visualize the predictions on training data
plot_confusion_matrix(cat_model,X_train,y_train)
plt.show()
```

##### OBSERVATION

- from `143893` orders , catboost model able to predict `86346` orders correctly

```python
# %Let's use confunsion matrix plot to Visualize the predictions on test data
plot_confusion_matrix(cat_model,X_test,y_test)
plt.show()
```

##### OBSERVATION

- from `35974` orders , catboost model able to predict `23570` orders correctly

#### BUILT-IN FEATURE IMPORTANCE

```python
# %Let's Visualize on the best contributed features
feat_importances = pd.Series(cat_model.feature_importances_, index=cols)
plt.figure(figsize=(15,10))
plt.title("BUILT-IN FEATURE IMPORTANCE PLOT")
feat_importances.nlargest(20).plot(kind='barh')
plt.show()
```

#### FEATURE IMPORTANCE COMPUTED WITH SHAP VALUES

```python
import shap

plt.figure(figsize=(15,10))
plt.title("COMPUTED SHAP VALUES FEATURE IMPORTANCE PLOT")
explainer = shap.TreeExplainer(cat_model)
shap_values = explainer.shap_values(X_test)
shap.summary_plot(shap_values, X_test, plot_type="bar")
plt.show()
```
